```r
which(apply(a, 2, var)== 0) # 38친구가 다 똑같은 값을 가짐 ㅠㅠ 제거해버릴 예정
b <- b[,c(-38)]
b_t <- b[, c(-53)];b_label <- b[, 53]
```

```r
p <- prcomp(b_t, scale = T, center = T)
p1 <- prcomp(a_t, scale = T, center = T)
```

```r
trainprc0 <- as.data.frame(cbind(as.matrix(a_t)%*% p1$rotation, a_label))
trainprc <- as.data.frame(cbind(as.matrix(b_t) %*% p$rotation, b_label))
model0 <- lm(b_label~., data = trainprc0)
model1 <- lm(b_label~., data = trainprc)
```

```r
head(trainprc)
```

```r
summary(model0) # 어느정도 성능 향상
```

```r
model1 <- lm(b_label ~.-PC51-PC52, data = trainprc)
summary(model1) # 변수를 2개나 뺐는데 동일한 성능
```
