```python
from konlpy.tag import Twitter
from gensim.models import Word2Vec
import pandas as pd
from tqdm import tqdm
import re
```

```python
data = pd.read_csv("tw_Cr.csv")
```

```python
data.iloc[:, 1].head
```

```python
ts = list(data.iloc[:, 1])
pat = re.compile(r"@+[A-z]+:")
re.split(pat, ts[0])
```

```python
texts = []
for i in ts:
    try:
        _, text = re.split(pat,i)
        texts.append(text)
    except:
        next
```

```python
len(texts)
```

```python
texts[0]
```

```python
twittwit = Twitter()
```

```python
token = []
for i in tqdm(texts):
    token.append(twittwit.nouns(i))
```

```python
token[0]
```

```python
model = Word2Vec(token, sg = 1, window = 5, size = 10, min_count = 5)
```

```python
model.wv.vectors.shape
```

```python
import copy
X = model.wv.vectors
W = list(model.wv.vocab.keys())
```

```python
from sklearn  import cluster
kmeans = cluster.KMeans(n_clusters = 10)
kmeans.fit(X)
```

```python
label = kmeans.labels_
```

```python
import numpy as np
```

```python
raw_data = np.c_[X, np.array(W), np.array(label)]
```

```python
United_data = pd.DataFrame(raw_data)
```
